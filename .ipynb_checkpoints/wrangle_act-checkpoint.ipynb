_kindly note that the block of code used below to query Twitter API for each tweet in the twitter archive was provided by Udacty in  [twitter_api.py](https://video.udacity-data.com/topher/2018/November/5be5fb4c_twitter-api/twitter-api.py)_

```python

# Query Twitter API for each tweet in the Twitter archive and save JSON in a text file
# These are hidden to comply with Twitter's API terms and conditions
consumer_key = 'XXXXXXX'
consumer_secret = 'XXXXXXX'
access_token = 'XXXXXXX'
access_secret = 'XXXXXXX'

auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit=True)

# NOTE TO STUDENT WITH MOBILE VERIFICATION ISSUES:
# df_1 is a DataFrame with the twitter_archive_enhanced.csv file. You may have to
# change line 17 to match the name of your DataFrame with twitter_archive_enhanced.csv
# NOTE TO REVIEWER: this student had mobile verification issues so the following
# Twitter API code was sent to this student from a Udacity instructor
# Tweet IDs for which to gather additional data via Twitter's API
tweet_ids = twitter_archive.tweet_id.values
len(tweet_ids)

# Query Twitter's API for JSON data for each tweet ID in the Twitter archive
count = 0
fails_dict = {}
start = timer()
# Save each tweet's returned JSON as a new line in a .txt file
with open('tweet_json.txt', 'w') as outfile:
    # This loop will likely take 20-30 minutes to run because of Twitter's rate limit
    for tweet_id in tweet_ids:
        count += 1
        print(str(count) + ": " + str(tweet_id))
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            print("Success")
            json.dump(tweet._json, outfile)
            outfile.write('\n')
        except tweepy.errors.TweepyException as e:
            print("Fail")
            fails_dict[tweet_id] = e
            pass
end = timer()
print(end - start)
print(fails_dict)
```

##### Define

Combine function and Bs4 to extract *__content__* and *__href__* from the html element and assign them to new columns `tweet_source` and `source_link` respectively. Drop `source` column using `.drop()`

```python
import bs4
```

##### Code

```python
def tweet_source(x):
    soup = bs4.BeautifulSoup(x, 'lxml')
    return soup.find('a').contents[0]

twitter_archive_clean['tweet_source'] = twitter_archive_clean['source'].apply(lambda y: tweet_source(y))
```

```python
def source_link(x):
    soup = bs4.BeautifulSoup(x, 'lxml')
    return soup.find('a')['href']
twitter_archive_clean['source_link'] = twitter_archive_clean['source'].apply(lambda x: source_link(x))
```

```python
# drop source column
twitter_archive_clean = twitter_archive_clean.drop(columns=['source'])
```

##### define

join the `doggo`, `floofer`, `pupper` and `puppo` as one column (`dog-stage`) using concat, create a function that records the stage(s) within the new column and then returns the values as the `dog_stage`. drop `doggo`, `floofer`, `pupper` and `puppo` columns

```python
# join doggo, floofer, pupper and puppo
twitter_archive_clean['dog_stage'] = twitter_archive_clean['doggo'].str.cat(twitter_archive_clean[['floofer', 'pupper', 'puppo']], sep='*')
```

```python
twitter_archive_clean['dog_stage'].value_counts()
```

```python
#create a function that records the stage(s) within the new column
def detectStage(x):
    if(x == 'None*None*pupper*None'):
        stage = 'pupper'
    elif(x == 'doggo*None*None*None'):
        stage = 'doggo'
    elif(x == 'None*None*None*puppo'):
        stage = 'puppo'
    elif(x == 'doggo*None*pupper*None'):
        stage = 'doggo/pupper'
    elif(x == 'None*floofer*None*None'):
        stage = 'floofer'
    elif(x == 'doggo*None*None*puppo'):
        stage = 'doggo/puppo'
    elif(x == 'doggo*floofer*None*None'):
        stage = 'doggo/floofer'
    else:
        stage = 'none'
    return stage
```

```python
#assign returnned value to dog_stage
twitter_archive_clean['dog_stage'] = twitter_archive_clean['dog_stage'].apply(lambda x: detectStage(x))
```

```python
#drop columns
twitter_archive_clean = twitter_archive_clean.drop(columns=['doggo', 'floofer', 'pupper', 'puppo'])
```

##### Define
create a dict that Maps existing laguage codes to their full name and create a function that applies it to the `lang` column.

##### Code

```python
# Mapping from language  abbreviation to full name
lang_abbrev = {'en': 'English',
                'und': 'Undetermined',
                'nl': 'Dutch',
                'in': 'Indonesian',
                'eu': 'Basque',
                'es': 'Spanish',
                'tl': 'tagalog',
                'et': 'Estonian'}

# Function to apply
def lang_code(x):
    if x['lang'] in lang_abbrev.keys():
        abbrev = lang_abbrev[x['lang']]
        return abbrev
    else:
        return x['lang']
    
tweets_clean['lang'] = tweets_clean.apply(lang_code, axis=1)
```
